## Contrastive divergence (CD)


```python
# eq(213) page 97, activation via sigmoid
# taking into account energy gap 2 for "spin" variables (-1,1)
#@jit(parallel=True)
def CD_step(v_in,wei,bias,details=False,POTTS=False):
    """
        Generates the state on the other layer: 
        Field "H" ->  activation "a" -> probability "p" -> Spins/bits v_out

        Either (v_in=x, wei=w) or (v_in=z, wei=w.T)

        details = True --> returns also probability (p) and activation (a) 

        POTTS means one-hot encoding (used only in hidden space)
    """
    # local "field"
    H = np.clip(np.dot(v_in.astype(np.float64), wei) + bias, a_min=-300, a_max=300)
    # "activation"
    a = np.exp(level_gap*H)
    n = np.shape(H)
    v_out = np.full(n, x_min, dtype=int) # initially, just a list on -1's or 0's
    if POTTS: # RBM with a single hidden unit = 1 (that is, "one-hot encoding" with L states)
        # p: state probability, normalized to 1 over all units=states
        p = a/np.sum(a)
        # F: cumulative probability
        F = np.cumsum(p)
        # pick a state "i" randomly with probability p[i]
        r = np.random.rand()
        i = 0
        while r>F[i]:
            i+=1
        v_out[i] = 1 # activate a single hidden unit
    else: # normal Ising RBM
        # p: local probability, normalized to 1 for each hidden unit
        p = a / (a + 1.)
        # at each position i, activate the 1's with local probability p[i]
        v_out[np.random.random_sample(n) < p] = 1 

    if details: return (v_out,p,a)
    else: return v_out
```

## Plot weights

```python
def plot_weights_bias(wE, bE, epoch, L, 
                      side=0,cols=0,thr=0,s=1.5, 
                      title=False, save=True,cmap="bwr"):
    '''
    Plot the weights of the RBM, one plot for each hidden unit.
    '''
    rows = int(np.ceil(L / cols))
    if rows==1: rows=2
    w=wE[epoch]
    b=bE[epoch]
    if side==0: side=int(sqrt(len(w)))
    if thr==0: thr=4
    plt.clf()
    fig, AX = plt.subplots(rows, cols+1, figsize=(s*(1+cols),s*rows))
    if title: fig.suptitle(f"epoch = {epoch}")
    k=1
    for i in range(rows):
        for j in range(cols):
            if rows==1: ax=AX[j+1]
            else: ax=AX[i,j+1]
            if k<=L:
                ax.imshow(w[:,k-1].reshape(side, side), cmap=cmap,vmin=-thr,vmax=thr)
                ax.set_xticks([])
                ax.set_yticks([])
                ax.set_title(f"hidden {k}")
            else: fig.delaxes(ax)
            k+=1
        if i>0:  fig.delaxes(AX[i,0])
    
    ax=AX[0,0];
    im=ax.imshow(b.reshape(side, side), cmap=cmap,vmin=-thr,vmax=thr)
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_title("bias")
    # colobar
    cbar_ax = fig.add_axes([0.14, 0.15, 0.024, 0.33])
    cbar = fig.colorbar(im, cax=cbar_ax)
    cbar.ax.tick_params(labelsize=12)
    
    S=0.3
    plt.subplots_adjust(hspace=S)

    if save: plt.savefig(f"./FIG/FRAME/RBM_{epoch}_w-a.png")

    plt.show()
    plt.close()
```

## Weights initialization 

```python
# initial bias of visible units, based on their average value in the dataset
# Hinton, "A Practical Guide to Training Restricted Boltzmann Machines"
def Hinton_bias_init(x):
    '''
    x ::shape of 'x'(data): (21770,784)
    xmean :: does the mean on 'data' for each column -> shape of xmean :(,784)
     xmin =0, xmax=1 if SPINS=FALSE, otherwise {-1,+1}
    since np.precision is set at 4 -> S = 1e-4
    level_gap :: is the difference in values between the max (1) and the min (x_min)
    np.clip :: avoid minimum maximum overflow, inf and -inf values.
    '''
    xmean=np.array(np.mean(x,axis=0))
    # remove values at extrema, to avoid divergences in the log's
    S = 1e-4
    x1,x2 = x_min+S,1-S 
    xmean[xmean<x1] = x1
    xmean[xmean>x2] = x2
    return (1/level_gap)*np.clip(log(xmean-x_min) - log(1-xmean),-300,300)
    
# range of each initial weight
# Glorot and Bengio, "Understanding the difficulty of training deep feedforward neural networks"
sigma = sqrt(4. / float(L + D))
```

### Log-Likelihood

```python
import numpy as np
import itertools as it

#This code has been optimized with the help of chatGPT 4o. It is now clear the importance of using numpy array instead of lists when dealing with repeated operations, and that appending to a vector is 
# not optimized, instead is better to initialize a scalar and accumulate to it. Also important to do all possible operations outside of loops.
#@jit(parallel=True)
def generate_configurations(L):
    """Generate all binary configurations of length L."""
    return np.array(list(it.product((0,1), repeat=L)), dtype=np.int64)
#@jit(parallel=True)
def compute_log_likelihood(partition,data, a, b, w):
    log_likelihoods = np.zeros(len(data))  # Preallocate array
    all_conf = generate_configurations(L)  # Compute binary configs once

    Hz_cache = np.array([H(z) for z in all_conf])  # Precompute H(z)

    for i, x in enumerate(data): 
        Z_x = 0.0
        for Hz, z in zip(Hz_cache, all_conf):
            E_xz = -np.dot(Hz, x) - np.dot(b, z)
            Z_x += np.exp(-E_xz)

        log_likelihoods[i] = np.log(Z_x) - partition

    return np.mean(log_likelihoods)

def H(z):  
    return a + np.dot(w, z)  

def G(z):  
    return np.prod(np.exp(b * z))
#@jit(parallel=True)

#@jit(forceobj=True, looplift=True)
def partition_function(configurations):
    #all_conf = generate_configurations(L)
    sum_values = 0.0  # Use scalar accumulation

    for z in configurations:
        Hz = H(z)
        produttoria_H = np.prod(1 + np.exp(Hz))
        sum_values += G(z) * produttoria_H

    return np.log(sum_values)



```

```python
# Setting up parameters
L = 2
Nt = 1
# random seed for reproducibility
np.random.seed(12345)

# nr of epochs
Nepoch = 150
# minibatches per epoch
Nmini = 20  
# minibatch size at initial epoch and final one
N_ini, N_fin = 10, 500
print(f"Nepoch={Nepoch}\nNmini={Nmini}")

# Define the k-fold cross validation parameters
k_folds = 5
kf = KFold(n_splits=k_folds, shuffle=True, random_state=42)

# Arrays to store results for each fold
fold_train_scores = []
fold_test_scores = []
fold_weights = []  # Store final weights from each fold
fold_visible_biases = []  # Store final visible biases from each fold
fold_hidden_biases = []  # Store final hidden biases from each fold

# For plotting learning curves
all_train_curves = []
all_test_curves = []
epochs_to_track = list(range(0, Nepoch+1, 5))  # Track every 5 epochs

print(f"Starting {k_folds}-fold cross validation")
print(f"Model parameters: L={L}, Nt={Nt}, Nepoch={Nepoch}")

# For each fold
for fold_idx, (train_index, test_index) in enumerate(kf.split(data)):
    print(f"\n{'='*50}")
    print(f"Training fold {fold_idx+1}/{k_folds}")
    print(f"{'='*50}")
    
    # Split data into training and validation sets
    train_data = data[train_index]
    test_data = data[test_index]
    Nd = len(train_data)
    print(f"Train set size: {len(train_data)}, Test set size: {len(test_data)}")
    
    # Initialize weights and biases with different seed for each fold
    np.random.seed(12345 + fold_idx)
    w = sigma * np.random.randn(D, L)
    a = Hinton_bias_init(train_data)
    b = np.zeros(L)
    
    # Generate all possible configurations for the log-likelihood calculation
    configurations = generate_configurations(L)
    
    # For RMSprop gradient descent
    if GRAD == "RMSprop":
        gw2, ga2, gb2 = np.zeros_like(w), np.zeros_like(a), np.zeros_like(b)
    
    # For tracking weights and biases history
    wE = np.zeros((Nepoch+1, D, L))
    aE = np.zeros((Nepoch+1, D))
    bE = np.zeros((Nepoch+1, L))
    wE[0], aE[0], bE[0] = np.copy(w), np.copy(a), np.copy(b)
    
    # Arrays to track metrics during training
    train_log_likelihood_history = []
    test_log_likelihood_history = []
    epochs_tracked = []
    
    # Calculate initial log-likelihood
    partition = partition_function(configurations)
    init_train_ll = compute_log_likelihood(partition, train_data, a=a, b=b, w=w)
    init_test_ll = compute_log_likelihood(partition, test_data, a=a, b=b, w=w)
    train_log_likelihood_history.append(init_train_ll)
    test_log_likelihood_history.append(init_test_ll)
    epochs_tracked.append(0)
    
    print(f"Initial Log-Likelihood - Train: {init_train_ll:.4f}, Test: {init_test_ll:.4f}")
    
    indices = np.arange(Nd).astype("int")
    
    # Begin training loop
    for epoch in range(1, 1+Nepoch):
        # q maps epochs to interval [0,1]
        q = (epoch-1.)/(Nepoch-1.) 
        # N, size of the mini batch
        N = int(N_ini + (N_fin-N_ini)*(q**2))
        # l_rate interpolates between initial and final value
        l_rate = l_rate_ini + (l_rate_fin-l_rate_ini)*q
        
        for mini in range(Nmini):
            # Initialize for averages in minibatch
            x_data, x_model = np.zeros(D), np.zeros(D)
            z_data, z_model = np.zeros(L), np.zeros(L)
            xz_data, xz_model = np.zeros((D,L)), np.zeros((D,L))
            pz = np.zeros(L)
            
            # Select minibatch from training data
            selected = np.random.choice(indices, N, replace=False)
            
            for k in range(N):
                # Use train_data instead of data
                x0 = train_data[selected[k]]
                # Positive CD phase
                z = CD_step(x0, w, b, POTTS=POTTS)
                x_data += x0
                z_data += z
                xz_data += np.outer(x0, z)
                
                # Fantasy particles
                zf = np.copy(z)
                # Contrastive divergence with Nt steps
                for t in range(Nt):
                    xf = CD_step(zf, w.T, a)
                    zf = CD_step(xf, w, b, POTTS=POTTS)
                    
                x_model += xf
                z_model += zf
                xz_model += np.outer(xf, zf)
                
                # Record probability of encoding in z-space, if POTTS
                if POTTS: 
                    pz[zf] += 1
            
            # Gradient calculation
            gw_d, gw_m = xz_data/N, xz_model/N
            ga_d, ga_m = x_data/N, x_model/N
            gb_d, gb_m = z_data/N, z_model/N
            gw = np.copy(gw_d - gw_m)
            ga = np.copy(ga_d - ga_m)
            gb = np.copy(gb_d - gb_m)
            
            # Parameter updates
            if GRAD == "RMSprop":
                # RMSprop gradient ascent
                gw2 = beta*gw2+(1-beta)*np.square(gw)
                ga2 = beta*ga2+(1-beta)*np.square(ga)
                gb2 = beta*gb2+(1-beta)*np.square(gb)
                w += l_rate*gw/sqrt(epsilon+gw2)
                a += l_rate*ga/sqrt(epsilon+ga2)
                b += l_rate*gb/sqrt(epsilon+gb2)
            else: 
                # Vanilla stochastic gradient ascent (SGD)
                w += l_rate*gw
                a += l_rate*ga
                b += l_rate*gb
                
            # LASSO regularization
            if gamma > 0.:
                w -= (gamma*l_rate)*sign(w)
                a -= (gamma*l_rate)*sign(a)
                b -= (gamma*l_rate)*sign(b)
        
        # Save weights and biases history
        wE[epoch] = np.copy(w)
        aE[epoch] = np.copy(a)
        bE[epoch] = np.copy(b)
        
        # Calculate log-likelihood every few epochs
        if epoch % 5 == 0 or epoch == Nepoch:
            partition = partition_function(configurations)
            train_ll = compute_log_likelihood(partition, train_data, a=a, b=b, w=w)
            test_ll = compute_log_likelihood(partition, test_data, a=a, b=b, w=w)
            
            train_log_likelihood_history.append(train_ll)
            test_log_likelihood_history.append(test_ll)
            epochs_tracked.append(epoch)
            
            print(f"Epoch {epoch}/{Nepoch}, N: {N}, rate: {l_rate:.6f}")
            print(f"Train Log-Likelihood: {train_ll:.4f}, Test Log-Likelihood: {test_ll:.4f}")
    
    # After training, store final results for this fold
    partition = partition_function(configurations)
    final_train_ll = compute_log_likelihood(partition, train_data, a=a, b=b, w=w)
    final_test_ll = compute_log_likelihood(partition, test_data, a=a, b=b, w=w)
    
    fold_train_scores.append(final_train_ll)
    fold_test_scores.append(final_test_ll)
    fold_weights.append(np.copy(w))
    fold_visible_biases.append(np.copy(a))
    fold_hidden_biases.append(np.copy(b))
    
    all_train_curves.append((epochs_tracked, train_log_likelihood_history))
    all_test_curves.append((epochs_tracked, test_log_likelihood_history))
    
    print(f"\nFold {fold_idx+1} Final Results:")
    print(f"Train Log-Likelihood: {final_train_ll:.4f}")
    print(f"Test Log-Likelihood: {final_test_ll:.4f}")
    
    # Plot learning curves for this fold
    plt.figure(figsize=(10, 6))
    plt.plot(epochs_tracked, train_log_likelihood_history, 'b-', label='Training Log-Likelihood')
    plt.plot(epochs_tracked, test_log_likelihood_history, 'r-', label='Test Log-Likelihood')
    plt.xlabel('Epochs')
    plt.ylabel('Log-Likelihood')
    plt.title(f'Fold {fold_idx+1} Learning Curves (L={L}, CD steps={Nt})')
    plt.legend()
    plt.grid(True)
    plt.savefig(f'img/fold{fold_idx+1}_learning_curves_L{L}_CD{Nt}.png')

# Calculate average scores
avg_train_score = sum(fold_train_scores) / len(fold_train_scores)
avg_test_score = sum(fold_test_scores) / len(fold_test_scores)

# Print summary results
print(f"\n{'='*50}")
print(f"K-fold Cross-Validation Summary (k={k_folds}):")
print(f"{'='*50}")

for i in range(k_folds):
    print(f"Fold {i+1}: Train LL = {fold_train_scores[i]:.4f}, Test LL = {fold_test_scores[i]:.4f}")

print(f"\nAverage Train Log-Likelihood: {avg_train_score:.4f}")
print(f"Average Test Log-Likelihood: {avg_test_score:.4f}")
print(f"Train/Test Gap: {avg_train_score - avg_test_score:.4f}")

# Calculate standard deviations
train_std = np.std(fold_train_scores)
test_std = np.std(fold_test_scores)
print(f"Train Log-Likelihood Std Dev: {train_std:.4f}")
print(f"Test Log-Likelihood Std Dev: {test_std:.4f}")

# Plot final results
plt.figure(figsize=(12, 6))

# Bar chart for final scores
plt.subplot(1, 2, 1)
x = np.arange(k_folds)
width = 0.35
plt.bar(x - width/2, fold_train_scores, width, label='Train')
plt.bar(x + width/2, fold_test_scores, width, label='Test')
plt.axhline(y=avg_train_score, color='b', linestyle='--', alpha=0.7, label=f'Avg Train: {avg_train_score:.4f}')
plt.axhline(y=avg_test_score, color='r', linestyle='--', alpha=0.7, label=f'Avg Test: {avg_test_score:.4f}')
plt.xlabel('Fold')
plt.ylabel('Log-Likelihood')
plt.title(f'Final Scores by Fold')
plt.xticks(x, [f'{i+1}' for i in range(k_folds)])
plt.legend()
plt.grid(True, alpha=0.3)

# Learning curves for all folds
plt.subplot(1, 2, 2)
for i, (epochs, train_ll) in enumerate(all_train_curves):
    plt.plot(epochs, train_ll, 'b-', alpha=0.3)
for i, (epochs, test_ll) in enumerate(all_test_curves):
    plt.plot(epochs, test_ll, 'r-', alpha=0.3)

# Calculate and plot average curves
common_epochs = all_train_curves[0][0]  # Assuming all folds track same epochs
avg_train_curve = np.zeros(len(common_epochs))
avg_test_curve = np.zeros(len(common_epochs))

for epochs, values in all_train_curves:
    avg_train_curve += np.array(values)
for epochs, values in all_test_curves:
    avg_test_curve += np.array(values)

avg_train_curve /= k_folds
avg_test_curve /= k_folds

plt.plot(common_epochs, avg_train_curve, 'b-', linewidth=2, label='Avg Train LL')
plt.plot(common_epochs, avg_test_curve, 'r-', linewidth=2, label='Avg Test LL')

plt.xlabel('Epochs')
plt.ylabel('Log-Likelihood')
plt.title('Learning Curves Across All Folds')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.savefig(f'img/cv_summary_L{L}_CD{Nt}_k{k_folds}.png')
plt.show()

print("\nCross-validation complete!")

# Return the best model based on test score
best_fold_idx = np.argmax(fold_test_scores)
best_test_score = fold_test_scores[best_fold_idx]
best_weights = fold_weights[best_fold_idx]
best_visible_biases = fold_visible_biases[best_fold_idx]
best_hidden_biases = fold_hidden_biases[best_fold_idx]

print(f"\nBest model from fold {best_fold_idx+1} with test log-likelihood: {best_test_score:.4f}")
print("This model's parameters can be used for further evaluation or deployment.")
```
